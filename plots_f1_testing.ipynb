```python
import pandas as pd
import numpy as np

file_path = 'initial_model_data/LSTM_default.txt'
file_path2 = 'initial_model_data/LSTM_0.01.txt'
file_path3 = 'initial_model_data/LSTM_0.002.txt'
file_path4 = 'initial_model_data/LSTM_0.0005.txt'
file_path5 = 'initial_model_data/LSTM_0.0001.txt'
file_path6 = 'initial_model_data/RNN_default.txt'
file_path7 = 'initial_model_data/TD_default.txt'
file_path8 = 'initial_model_data/TD_default2.txt'
file_path9 = 'initial_model_data/probabilities.txt'






df = pd.read_csv(file_path, sep=';', header=None)
df2 = pd.read_csv(file_path2, sep=';', header=None)
df3 = pd.read_csv(file_path3, sep=';', header=None)
df4 = pd.read_csv(file_path4, sep=';', header=None)
df5 = pd.read_csv(file_path5, sep=';', header=None)
df6 = pd.read_csv(file_path6, sep=';', header=None)
df7 = pd.read_csv(file_path7, sep=';', header=None)
df8 = pd.read_csv(file_path8, sep=';', header=None)
df9 = pd.read_csv(file_path9, sep=';', header=None)


```

```python
# # data = np.row_stack((thresholds, train_success_list, train_failed_list, train_total_list, test_success_list, test_failed_list, test_total_list, test_attack_count_perc))
import matplotlib.pyplot as plt


legend_fontsize = 14
axis_fontsize = 16
tick_fontsize = 12

# First plot
plt.figure(figsize=(10, 6))
plt.plot(df.iloc[0].to_numpy(), df.iloc[2].to_numpy(), linewidth=3, color='#8B1E3F', label='Train Failed', linestyle='--')
plt.plot(df.iloc[0].to_numpy(), df.iloc[1].to_numpy(), linewidth=3, color='#386641', label='Train Successful', linestyle='--')
plt.plot(df.iloc[0].to_numpy(), df.iloc[5].to_numpy(), linewidth=5, color='#FF6663', label='Test Failed')
plt.plot(df.iloc[0].to_numpy(), df.iloc[4].to_numpy(), linewidth=5, color='#679436', label='Test Successful')
plt.xlabel('Binary Threshold', fontsize=axis_fontsize)
plt.ylabel('Number of Attacks Executed', fontsize=axis_fontsize)
plt.legend(fontsize=legend_fontsize)
plt.grid(True)

plt.tick_params(axis='both', which='major', labelsize=tick_fontsize)
# plt.savefig("decay_learning_rate_0.001_%_chance.png")


width = 0.01

# Second plot
fig, ax1 = plt.subplots(figsize=(10, 6))
# ax1.bar(df.iloc[0].to_numpy() - width/2, df.iloc[4].to_numpy()+df.iloc[5].to_numpy(), width=width, color='#679436', label='LSTM test results', alpha=0.7, zorder=1)
ax1.bar(df.iloc[0].to_numpy(), df.iloc[4].to_numpy()+df.iloc[5].to_numpy(), label='Test Attack Count', width=0.01, color='#A6ECE0', zorder=1)


ax1.set_xlabel('Binary Threshold', fontsize=axis_fontsize)
ax1.set_ylabel('Number of Attacks Executed', fontsize=axis_fontsize)
ax1.legend(loc='center right', fontsize=legend_fontsize)

ax1.tick_params(axis='both', which='major', labelsize=tick_fontsize)

ax2 = ax1.twinx()
ax2.grid(True, zorder=0)

ax2.plot(df.iloc[0].to_numpy(), df.iloc[3].to_numpy()/100, label='Train', linewidth=3, color='#386641', zorder=2, linestyle='--')
ax2.plot(df.iloc[0].to_numpy(), df.iloc[6].to_numpy()/100, label='Test', linewidth=4, color='#679436', zorder=2)
# ax2.plot(df.iloc[0].to_numpy(), df.iloc[3].to_numpy()/100, label='Train', linewidth=4, color='#679436', zorder=2)
# ax2.plot(df.iloc[0].to_numpy(), df.iloc[6].to_numpy()/100, label='Test', linewidth=4, color='#DD6E42', zorder=2)

ax2.set_yticks(np.arange(0.1, 1.1, 0.1))
ax2.set_yticklabels(['{:.1f}'.format(x) for x in np.arange(0.1, 1.1, 0.1)])
ax2.set_ylabel('Attack Success Probability', fontsize=axis_fontsize)
ax2.tick_params(axis='y', which='major', labelsize=tick_fontsize)


plt.legend(loc='upper right', fontsize=legend_fontsize)

plt.show()
```
